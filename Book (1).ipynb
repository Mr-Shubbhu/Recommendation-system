```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

# Cosine Similarity

```python
Book_sim = 1 - pairwise_distances( Book_pivot.values,metric='cosine')
```

```python
Book_sim
```

```python
book_sim_df = pd.DataFrame(Book_sim)
```

```python
book_sim_df.shape
```

```python
book_sim_df
```

```python
book_sim_df.index = Book_pivot.index
book_sim_df.columns = Book_pivot.index
```

```python
np.fill_diagonal(Book_sim, 0)
book_sim_df.iloc[0:5, 0:5]
```

```python
def get_reco(customer_id):
    
    try :
        tem=list(book_sim_df.sort_values([customer_id],ascending=False).head(100).index)
        book_list=list(Book[Book["User_ID"].isin(tem)]["Title"])
  
        return set(book_list)-set(Book[Book["User_ID"]==customer_id]["Title"])
    except : 
        return "the user id is not available"
```
